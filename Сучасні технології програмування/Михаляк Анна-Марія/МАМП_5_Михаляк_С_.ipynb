```python
%%writefile Program.cs
using System;
using System.IO;
using System.Text;

class Program
{
    const string FileName = "notes.txt";
    const int Shift = 3;

    static void Main()
    {
        int choice;
        do
        {
            PrintMenu();
            Console.Write(" : ");
            if (!int.TryParse(Console.ReadLine(), out choice))
            {
                Console.WriteLine(" .");
                continue;
            }

            switch (choice)
            {
                case 1:
                    CreateNote();
                    break;
                case 2:
                    ShowNotes();
                    break;
                case 3:
                    Console.WriteLine("...");
                    break;
                default:
                    Console.WriteLine(" .");
                    break;
            }
        } while (choice != 3);
    }

    static void PrintMenu()
    {
        Console.WriteLine("\n===    ===");
        Console.WriteLine("1.  ");
        Console.WriteLine("2.   ");
        Console.WriteLine("3. ");
    }

    static void CreateNote()
    {
        Console.Write("  : ");
        string text = Console.ReadLine() ?? "";
        string encrypted = Encrypt(text);

        File.AppendAllText(FileName, encrypted + Environment.NewLine, Encoding.UTF8);
        Console.WriteLine("  (  ).");
    }

    static void ShowNotes()
    {
        if (!File.Exists(FileName))
        {
            Console.WriteLine("  .");
            return;
        }

        string[] lines = File.ReadAllLines(FileName, Encoding.UTF8);
        Console.WriteLine("\n===   () ===");
        foreach (var encrypted in lines)
        {
            string decrypted = Decrypt(encrypted);
            Console.WriteLine("- " + decrypted);
        }
    }

    static string Encrypt(string input)
    {
        var sb = new StringBuilder();
        foreach (char c in input)
        {
            sb.Append((char)(c + Shift));
        }
        return sb.ToString();
    }

    static string Decrypt(string input)
    {
        var sb = new StringBuilder();
        foreach (char c in input)
        {
            sb.Append((char)(c - Shift));
        }
        return sb.ToString();
    }
}
```
